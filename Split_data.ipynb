```python
import os
import shutil
import random
from pathlib import Path

# ============================================================
# CẤU HÌNH — chỉ cần sửa 2 dòng này
# ============================================================
SRC_DIR = Path(r"F:\Project\Food_Classification/data")   # thư mục chứa 101 folder class
DST_DIR = Path(r"F:\Project\Food_Classification/data_split")  # output

TRAIN_RATIO = 0.70
VAL_RATIO   = 0.20
TEST_RATIO  = 0.10
SEED        = 42
# ============================================================

random.seed(SEED)
IMG_EXTS = {".jpg", ".jpeg", ".png", ".webp", ".bmp"}

def split_dataset():
    class_dirs = [d for d in SRC_DIR.iterdir() if d.is_dir()]
    print(f"Tìm thấy {len(class_dirs)} class\n")

    stats = []

    for cls_dir in sorted(class_dirs):
        images = [f for f in cls_dir.iterdir() if f.suffix.lower() in IMG_EXTS]
        random.shuffle(images)

        n = len(images)
        n_train = int(n * TRAIN_RATIO)
        n_val   = int(n * VAL_RATIO)
        # phần còn lại vào test (tránh mất ảnh do làm tròn)
        n_test  = n - n_train - n_val

        splits = {
            "train": images[:n_train],
            "val":   images[n_train : n_train + n_val],
            "test":  images[n_train + n_val :],
        }

        for split_name, split_imgs in splits.items():
            out_dir = DST_DIR / split_name / cls_dir.name
            out_dir.mkdir(parents=True, exist_ok=True)
            for img in split_imgs:
                shutil.copy2(img, out_dir / img.name)

        stats.append((cls_dir.name, n, n_train, n_val, n_test))
        print(f"  {cls_dir.name:<30} total={n:>4}  train={n_train:>3}  val={n_val:>3}  test={n_test:>3}")

    # tổng kết
    total = sum(s[1] for s in stats)
    print(f"\nTổng: {total} ảnh / {len(stats)} class")
    print(f"Output: {DST_DIR.resolve()}")

split_dataset()
```
